```python
import numpy as np
import scipy as sp
import constants as c
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from simu import Simu
```

```python
r = 2.2
Omega_mid = 1.0/r**3 #Keplerian angular velocity at the midplane
z = np.linspace(0,1,20)
```

Takeuchi & Lin (2002): Eq. (15) approximation of the Keplerian angular velocity above the midplane

$$\Omega_K(r,z)\approx\Omega_{K,mid}\cdot\bigg(1-\frac{3}{4}\frac{z^2}{r^2}\bigg)$$
with 
$$\Omega_{K,mid}=\sqrt{\frac{GM}{r^3}}$$

```python
Omega1 = Omega_mid*(1.-3./4.*(z/r)**2) #Takeushi & Lin 2002, Eq 15.
```

horizontal balance of zentripedal acceleration and gravitational acceleration

$$\Omega^2r=\frac{GM}{r_s}\frac{r}{r_s}$$
with $r_s$ being the sphgerical radius $r_s^2=r^2+z^2$.
solving for $\Omega$ gives
$$\Omega_K(r,z)=\Omega_{K,mid}\bigg( 1+\frac{z^2}{r^2}\bigg)^{-3/4}$$

```python
rs = np.sqrt(r**2+z**2)
Omega2 = Omega_mid*(1.+(z/r)**2)**(-3./4.)
```

```python
plt.figure()
plt.plot(z,Omega1,label='approx.')
plt.plot(z,Omega2,label='exact')
plt.ylabel(r'$\Omega$')
plt.xlabel('z')
plt.legend()
plt.show()
```
